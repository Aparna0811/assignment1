### 1) displaying the grade

```python
marks=int(input("enter your marks"))
if marks>90:
    print("your grade is :A")
elif marks>80 and marks<=90:
    print("your grade is :B")
elif marks>=60 and marks<=80:
    print("your grade is :C") 
else:
    print("your grade is :D")
```

### 2) displaying the tax

```python
cost=int(input("enter your cost"))
if cost>100000:
    print("you have to pay the tax:15%")
elif cost>50000 and cost<=100000:
    print("you have to pay the tax:10%")
else:
    print("you have to pay the tax:5%")
        
    
```

### 3) displaying the monuments

```python
city=input("enter your city")
if  city=="Delhi":
    print("monuments of the Delhi city is:Red Fort")
elif city=="Agra":
    print("monuments of the Agra city is:Tah Mahal")
elif city=="Jaipur":
    print("monuments of the Jaipur city is:Jal Mahal")  
else:
    print("wrong choice")
```
